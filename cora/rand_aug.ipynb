```python
import sys
from pathlib import Path
notebook_dir = Path().resolve()
ROOT = notebook_dir.parent

sys.path.append(str(ROOT))

import torch
from torch_geometric.datasets import Planetoid
from src.models.mlp import MLP
from src.trainMethoden import train_graphcl
from src.models.svm import evaluate_with_svm
from src.plotting import save_results
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("device: ", device)
dataset = Planetoid(root='~/datasets/Cora', name='Cora')
data = dataset[0]
data= data.to(device)
print(data)
```

```python
# contrastive learning with random augmentations
# evaluate with MLP and SVM, save results for all runs
 
encoders = ["gcn", "sage", "gat", "gin"]
mode = "Rand-Aug"
num_runs = 5
all_run_accs = []
all_run_losses = []
all_run_embeddings = []
all_run_svm = []

for run in range(num_runs):
    print(f"\nRun {run+1}/{num_runs}")

    models, embeddings, loss_history = train_graphcl(
        data=data,
        encoders=encoders,
        device=device,
        mode=mode,
        contrastive_epochs=100, 
        batch_size=4096,
        tau=0.4,
        lr=1e-3,
        drop_feat=0.3,
        drop_edge=0.2, 
        noise_std=0.05,
        visualize=True,
        use_batching=False
    )

    _, best_acc = MLP.evaluate_with_mlp(
        embeddings=embeddings,
        data=data,
        device=device,
        mlp_epochs=120
    )

    all_run_accs.append(best_acc)
    all_run_losses.append(loss_history)
    all_run_embeddings.append({enc: emb.detach().cpu() for enc, emb in embeddings.items()})

    svm_results = evaluate_with_svm(
        embeddings=embeddings,
        data=data,
        device=device
    )
    all_run_svm.append(svm_results)
```

```python
# saving average results and std deviations

results_dir = Path("/Users/emely/Uni/Thesis/Thesis/cora/results")
results_dir.mkdir(parents=True, exist_ok=True)

lossfile = results_dir / "loss_histories.pt"
emb_file = results_dir / "embeddings.pt"
acc_file = results_dir / "best_accs.pt"
svm_file = results_dir / "svm_results.pt"

filenames = {'loss': lossfile, 'emb': emb_file, 'acc': acc_file, 'svm': svm_file}

avg_losses, avg_embeddings, avg_acc, std_acc, avg_svm, std_svm = save_results(
    mode=mode,
    encoders=encoders,
    all_run_losses=all_run_losses,
    all_run_embeddings=all_run_embeddings,
    all_run_accs=all_run_accs,
    all_run_svm=all_run_svm,
    filenames=filenames
)

print("Average Accuracies:", avg_acc)
print("Standard Deviations:", std_acc)
print("Average SVM Accuracies:", avg_svm)
print("Standard Deviations SVM:", std_svm)
```
